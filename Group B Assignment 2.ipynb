### Outliers in so2

```python
plt.boxplot(data5['so2'])
```

```python
q1 = np.quantile(data5['so2'],0.25)
q3 = np.quantile(data5['so2'],0.75)

iqr = q3 - q1
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)

outlier = []
for x in data5['so2']:
    if((x>upper_bound) or (x<lower_bound)):
        outlier.append(x)
print(outlier)
```

```python
new_data1 = data5[(data5['so2'] <= upper_bound) & (data5['so2'] >= lower_bound)]
```

```python
plt.boxplot(new_data1['so2'])
```
