```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("heart.csv")
data
```

### BARPLOT

```python
sns.barplot(data=data,x="sex",y="age")
```

### COUNTPLOT

```python
plt.figure(figsize = (10,4) ,dpi = 100)
color_palette = sns.color_palette("Accent_r")
sns.set_palette(color_palette)
sns.countplot(x = "slope", data=data)
```

```python
plt.figure(figsize = (10,4), dpi =100)
color_palette = sns.color_palette("Accent_r")
sns.set_palette(color_palette)
sns.countplot(x = "age" , data=data)
```

```python
plt.figure(figsize = (10,4) , dpi = 100)
color_palette = sns.color_palette("Accent_r")
sns.set_palette(color_palette)
sns.countplot(x = "sex" , data=data)
```

### PIECHART

```python
plt.figure(figsize = (10, 4), dpi = 100)
grp = dict(data.groupby('thal').groups)
m = {}
for key, val in grp.items():
    if key in m:
        m[key] += len(val)
    else:
        m[key] = len(val)
plt.title("Distribution of thal")
plt.pie(m.values(), labels = m.keys())
plt.show()
```

```python
plt.figure(figsize = (10,4) , dpi=100)
grp = dict(data.groupby('sex').groups)
m = {}
for key, val in grp.items():
    if key in m:
        m[key] = len(val)
    else:
        m[key] = len(val)
plt.title("Distribution of Gender")
plt.pie(m.values(), labels = m.keys())
plt.show()
```
